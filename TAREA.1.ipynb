8. Escribir un programa que pida al usuario un número entero y muestre en pantalla un tríangulo rectángulo.


```
def triangulo(n):
    """
    for i in range(1,n+1) 
    "*"*i 
    [] 
    "\n".join() 
    """
    return "\n".join(["*"*i for i in range(1,n+1)])
 
numero=int(input("Dame un número: "))
print(triangulo(numero))
print("Fin del programa")
```

9. Escribir una función que recibe un número entero positivo y devuelva su factorial.


```
n = int (input("Dame el número del cual quieres saber el factorial:" ))
def factorial(n):
  if n==0:
    return 1
  else:
    return n * factorial(n-1)
print("El nuúmero factorial de:",n,"es:",factorial(n))
print("Fin del programa")
```
